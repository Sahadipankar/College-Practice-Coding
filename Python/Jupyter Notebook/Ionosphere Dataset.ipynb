```python
# Pairplot
sns.pairplot(data[["V1", "V2", "V3", "V4", "Class"]], hue='Class', diag_kind='kde')
plt.suptitle('Pairplot of Features', y=1.02)
plt.show()
```

```python
# Boxplot
plt.figure(figsize=(15, 10))
sns.boxplot(data=data.drop(columns=['Class']))
plt.title('Boxplot of Features')
plt.xticks(rotation=90)
plt.show()
```

```python
# Heatmap
plt.figure(figsize=(15, 10))
correlation_matrix = data.drop(columns=['Class']).corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Heatmap of Feature Correlations')
plt.show()

```

```python
plt.figure(figsize=(11, 7))
mean_values.plot(kind='bar', ax=plt.gca())
plt.title('Average Feature Values by Class')
plt.xlabel('Class')
plt.ylabel('Average Value')
plt.xticks(rotation=0)
plt.legend(loc='upper right')
plt.tight_layout()
plt.show()
```

```python
# Histogram
plt.figure(figsize=(15, 20))
data.drop(columns=['Class']).hist(bins=30, figsize=(15, 10))
plt.suptitle('Histograms of Features', y=1)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])  # Adjusting layout to add space for titles
plt.show()
```
