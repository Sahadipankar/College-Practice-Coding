```python
df = pd.read_csv(r"/content/drive/MyDrive/Colab Notebooks/Files/iris.csv", usecols=['setosa', 'versicolor'])

df
```

```python
df["setosa"].mean()
df["versicolor"].mean()

print("Mean of setosa:", df["setosa"].mean())
print("Mean of versicolor:", df["versicolor"].mean())
```

```python
df["setosa"].median()
df["versicolor"].median()

print("Median of setosa:", df["setosa"].median())
print("Median of versicolor:", df["versicolor"].median())
```

```python
df["setosa"].mode()[0]
df["versicolor"].mode()[0]

print("Mode of setosa:", df["setosa"].mode()[0])
print("Mode of versicolor:", df["versicolor"].mode()[0])
```
