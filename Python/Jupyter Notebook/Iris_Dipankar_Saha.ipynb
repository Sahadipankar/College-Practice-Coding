```python
# List all columns containing numeric values
numeric_columns = data.select_dtypes(include=[np.number]).columns.tolist()

print("Numeric columns:", numeric_columns)
```

```python
# Set display option to show numbers in standard decimal format
pd.set_option('display.float_format', lambda x: '%.2f' % x)

# Calculate the mean for all columns again
mean_values = data.mean(numeric_only=True)
print("Mean values for each column:\n")
print(mean_values)
```

```python
# Calculate the median for all columns
median_values = data.median(numeric_only=True)
print("Median values for each column:\n")

print(median_values)
```

```python
sns.pairplot(data[['150', '4', 'setosa', 'versicolor', 'virginica']])
plt.suptitle('Pairplot of Features', y=1.02)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=data[['150', '4', 'setosa', 'versicolor', 'virginica']])
plt.title('Boxplot of Features')
plt.show()
```

```python
corr_data = data[["150", "4", "setosa", "versicolor", "virginica"]].corr()
print(corr_data)

sns.heatmap(corr_data, annot=True)
plt.title('Heatmap of Correlation Matrix')
plt.show()
```

```python
avg_measurements = data[['setosa', 'versicolor', 'virginica']].mean()
plt.figure(figsize=(10, 6))
sns.barplot(x=avg_measurements.index, y=avg_measurements.values)
plt.title('Barplot of Average Measurements by Species')
plt.ylabel('Average Measurement')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['150'], bins=30, kde=False)
plt.title('Histogram')
plt.xlabel('150')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.displot(data['4'], bins=30, kde=True)
plt.title('Distribution Plot')
plt.xlabel('4')
plt.ylabel('Density')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data['150'], y=data['4'], hue=data['setosa'] + data['versicolor'] + data['virginica'])
plt.title('Scatterplot of 150 vs. 4')
plt.xlabel('150')
plt.ylabel('4')
plt.legend(title='Species')
plt.show()
```
