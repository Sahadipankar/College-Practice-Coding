```python
# Plot histograms for all columns
data["AGE"].hist()
plt.title("Age Distribution", fontsize=20) 
plt.xlabel("Age") 
plt.ylabel("Frequency")
plt.show()
```

```python
data.plot(kind='scatter', x='AGE', y='TAX')

plt.title("Age vs Tax")  

plt.xlabel("AGE")  

plt.ylabel("TAX")  

plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=data[['CRIM', 'ZN', 'INDUS', 'NOX', 'RM']])
plt.title("Box plot of selected features")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.displot(data, x="MEDV", kde=True)
plt.title("Distribution plot of MEDV")
plt.show()
```

```python
corr_data = data[["AGE", "RAD"]]

corr_data.head()
```

```python
sns.heatmap(corr_data, annot=True, fmt=".2f", cmap='coolwarm')
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
selected_columns = data[['AGE', 'TAX']]

sns.pairplot(selected_columns)
plt.suptitle("Pairplot of 'AGE' and 'TAX'", fontsize=20, y=1.06)  # Set the title for the pairplot
plt.show()
```

```python
# Pairplot to visualize pairwise relationships
sns.pairplot(data)
plt.suptitle('Pairplot of Boston House Prices', y=1.02)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='TAX', y='AGE', data=data, estimator=np.mean, errorbar=None)
plt.xticks(rotation=90)
plt.title('Average Tax')
plt.show()
```
