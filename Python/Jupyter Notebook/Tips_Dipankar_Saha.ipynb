```python
male_count = data["sex"].value_counts().get("Male",0)
print(f"Number of males: {male_count}")
```

```python
# Filter the data for males
male_data = data[data["sex"] == "Male"]

# Count the number of males who have given tips
male_tip_count = male_data["tip"].count()

# Sum of tips given by males
total_tips_by_males = male_data["tip"].sum()

print(f"Number of males who have given tips: {male_tip_count}")
print(f"Total amount of tips given by males: ${total_tips_by_males:.2f}")
```

```python
# Find the highest tip given by males
highest_tip_by_male = male_data["tip"].max()

# Find the lowest tip given by males
lowest_tip_by_male = male_data["tip"].min()

print(f"Highest tip given by a male: ${highest_tip_by_male:.2f}")
print(f"Lowest tip given by a male: ${lowest_tip_by_male:.2f}")
```
