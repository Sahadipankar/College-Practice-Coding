Dataset with missing values. Sometimes, the dataset may contain some missing values. These missing values  must  be  detected and the corresponding rows can  be  deleted  from the   table. The missing values in the  table  are   marked  by  the    NaN value. There are several functions  to  remove or  fill  those  values.  

Sometimes the dataset might contain the  unwanted values  which are not at all related to the  other  values saved into the  table.
These are  called the   noises(These data will  not show any sort of  relevance or relation to the  other  data ) or also called as the  spurious values. These values  must   be   also   identified and  removed.

Example  for the  noise  is that  if  the  age  is  stored as  negative  in the  dataset  then  it  will  be called as the   noise.

dropna():- This function is  used to drop the  row containing  the   null values from the   dataset.

Attribute with dropna():-  One attribute inplace = true can be  used  with the dropna():- If  it gets  any  NULLvalue  inside  dataset , it   return  the true  value.

fillna() :- This simply fills the  missing value into the  dataset .
Sometimes the  particular  column value  has  been  chosen where the fillna() function  can be applied.

Sometimes some statistical function is  used    to  compute the  value and the  value  can be   used  to  replace the missing  value with fillna().

The statistical functions to  be applied are the   mean(), median() and  mode()


```python
import scipy.stats

x = np.arange(10,15)
y = np.array([11,15,16,20,25])
print(x)
print(y)
```

```python
r = np.corrcoef(x, y)

r
```

```python
print(r[0, 0])
print(r[0, 1])
print(r[1, 0])
print(r[1, 1])
```

```python
r1 , p1 = scipy.stats.pearsonr(x,y)
print(r1 , p1)
r2 ,p2 = scipy.stats.spearmanr(x,y)
print(r2 , p2 ) 
r3 ,p3= scipy.stats.kendalltau(x,y)
print(r3 ,p3)
```
