```python
pd.set_option('display.float_format', lambda x: '%.2f' % x)

mean_values = data.mean(numeric_only=True)
print("Mean values for each column:\n")
print(mean_values)
```

```python
median_values = data.median(numeric_only=True)
print("Median values for each column:\n")

print(median_values)
```

```python
corr_data = data[["Length", "Diameter", "Height", "Whole weight"]].corr()
print(corr_data)
```

```python
# Pairplot
sns.pairplot(data.drop(columns=['Sex']), diag_kind='kde')
plt.suptitle('Pairplot of Selected Features', y=1.02)
plt.show()
```

```python
# Boxplot
plt.figure(figsize=(10, 8))
sns.boxplot(data=data.drop(columns=['Rings']))
plt.title('Boxplot of Selected Features')
plt.xticks(rotation=45)
plt.show()
```

```python
# Heatmap
plt.figure(figsize=(8, 6))
correlation_matrix = filtered_data.drop(columns=['Sex']).corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Heatmap of Selected Feature Correlations')
plt.show()
```

```python
# Barplot
plt.figure(figsize=(10, 8))
sns.barplot(x=mean_values.index, y=mean_values.values)
plt.title('Barplot of Average Feature Values')
plt.xlabel('Feature')
plt.ylabel('Average Value')
plt.xticks(rotation=45)
plt.show()
```

```python
# Histogram
plt.figure(figsize=(8, 6))
sns.histplot(data['Length'], bins=10, kde=False)
plt.title('Histogram of Length')
plt.xlabel('Length')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.displot(data['Diameter'], bins=10, kde=True)
plt.title('Distribution Plot of Diameter')
plt.xlabel('Diameter')
plt.ylabel('Density')
plt.show()
```

```python
# Scatterplot 
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Length', y='Whole weight', hue='Sex', data=data)
plt.title('Scatterplot of Length vs Diameter')
plt.xlabel('Length')
plt.ylabel('Whole weight')
plt.show()
```
